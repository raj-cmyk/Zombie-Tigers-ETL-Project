```python
# Dependencies
import pandas as pd
from config import username, password
from sqlalchemy import create_engine
```

# Connect to postgres

```python
#connect database
#connection_string = key
#engine = create_engine(f'postgresql://{connection_string}')
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/NPS')
```

```python
engine.table_names()
```

```python
pd.read_sql_query ('select * from species', con=engine).head()
```

```python
parkcount=pd.read_sql_query('select count(*) as count, park_name from species group by park_name',
                 con=engine)
parkcount.head()
```

```python
acre=pd.read_sql_query('Select count (*) as count, p.acres, p.park_name from park p\
                  JOIN species s on s.park_name=p.park_name\
                  group by p.acres, p.park_name',
                 con=engine)
acre
```

```python
category=pd.read_sql_query("select count(*) as count, park_name, category from species\
                           WHERE park_name='Rocky Mountain National Park'\
                           group by park_name, category",
                 con=engine)
category
```

```python
statecount=pd.read_sql_query('select count(*) as count, state from park p\
                  JOIN species s on s.park_name=p.park_name\
                  group by state',
                 con=engine)
statecount
```

```python
pd.read_sql_query ("select * from species\
                    WHERE park_name='Rocky Mountain National Park'\
                   AND category ='Crab/Lobster/Shrimp'", con=engine).head()
```
