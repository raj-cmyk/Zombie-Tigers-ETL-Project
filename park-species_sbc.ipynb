```python
pd.read_sql_query ('select * from species', con=ak.pg_on).head()
```

```python
parkcount=pd.read_sql_query('select count(*) as count, park_name from species group by park_name',
                 con=ak.pg_on)
parkcount.head()
```

```python
acre=pd.read_sql_query('Select count (*) as count, p.acres, p.park_name from park p\
                  JOIN species s on s.park_name=p.park_name\
                  group by p.acres, p.park_name',
                 con=ak.pg_on)
acre
```

```python
category=pd.read_sql_query("select count(*) as count, park_name, category from species\
                           WHERE park_name='Rocky Mountain National Park'\
                           group by park_name, category",
                 con=ak.pg_on)
category
```

```python
statecount=pd.read_sql_query('select count(*) as count, state from park p\
                  JOIN species s on s.park_name=p.park_name\
                  group by state',
                 con=ak.pg_on)
statecount
```
